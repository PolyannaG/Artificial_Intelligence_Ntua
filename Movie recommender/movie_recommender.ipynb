##Ερώτημα 1β

Έχοντας λοιπόν δημιουργήσει τον κόσμο του προβλήματος δημιουργούμε ένα σύνολο από απλούς κανόνες οι οποίοι θα βρίσκουν όλες τις ταινίες με:
1.	Κοινό θέμα (κάποιες λέξεις σχετικά με το genre κοινές)
2.	Αρκετά κοινό θέμα (κάποιες λιγότερες λέξεις σχετικά με το genre κοινές π.χ. 3)
3.	Σχετικά κοινό θέμα (λίγες λέξεις σχετικές με το genre κοινές π.χ. 1)
4.	Κοινός σκηνοθέτης
5.	Ακριβώς ίδια πλοκή (κάποιες λέξεις κλειδιά της πλοκής κοινές)
6.	Σχετικά ίδια πλοκή (κάποιες λιγότερες λέξεις κλειδιά κοινές)
7.	Ίδιους τους βασικούς ηθοποιούς (και τους 3)
8.	Αρκετά ίδιους βασικούς ηθοποιούς (ορισμένους βασικούς ηθοποιούς κοινούς π.χ. 2)
9.	Σχετικά ίδιους ηθοποιούς (π.χ. 1 από τους 3)
10.	Ίδια γλώσσα
11.	Είναι έγχρωμες ή ασπρόμαυρες
12. Κοινό studio παραγωγής
13. Κοινή χώρα παραγωγής 
14. Ίδια δεκαετία 

Προσθέτουμε επιπλέον ορισμένους κανόνες, σε περιπτώσεις ομοιότητας η οποία λαμβάνει κάποια κλίμακα (π.χ. ένα, δύο, τρία κλπ κοινά genres), ώστε να έχουμε πιο μεγάλη πληρότητα, καθώς και ορισμένους κανόνες που σχετίζονται με άλλα δοσμένα χαρακτηρηστικά, π.χ. vote_average, οι οποίοι δεν ζητώνται παραπάνω:

> Κανόνας 11  (black_and_white)

Στον κανόνα αυτό χρησιμοποιούμε ορισμένους βοηθητικούς κανόνες. Ο κανόνας find_black βρίσκει όλα τα ζεύγη ταινιών που είναι ασπρόμαυρες, ο κανόνας is_black_and_white ελέγχει αν μία ταινία είναι ασπρόμαυρη, και ο κανόνας find_color βρίσκει για μία δοσμένη έγχρωμη ταινία όλες τις υπόλοιπες εκτός αυτής οι οποίες είναι έγχρωμες. Τέλος, ο κανόνας black_and_white βρίσκει τις ταινίες που είναι όμοιες, δηλαδή είναι και οι δύο έγχρωμες ή και οι δύο ασπρόμαυρες.

```
find_black(Movie_1,Movie_2):-
  plot_keyword(Movie_1,'black and white'),
  plot_keyword(Movie_2,'black and white'), 
  Movie_1 \= Movie_2.
 
is_black_and_white(Movie):-
  plot_keyword(Movie,'black and white').


find_color(Movie_1,Movie_2):- 
  plot_keyword(Movie_2, Y), ( is_black_and_white(Movie_2) -> false ;
  Movie_1 \= Movie_2
  ).
  
black_and_white(Movie_1,Movie_2):- (
    is_black_and_white(Movie_1) -> find_black(Movie_1,Movie_2);
    find_color(Movie_1,Movie_2)
    ).
```

````python
def clean_text(text):                   # συνάρτηση που επεξεργάζειται strings πραγματοποιώντας ορισμένες απαραίτητες αντικαταστάσεις χαρακτήρων
  text = text.replace(u'\xa0', u'')
  text = text.replace(u"'", u'')
  return text
````

````python
import ast
#create World
#Ορίζουμε τον κόσμο μας
prolog = Prolog()

#Για κάθε row του πίνακα φτιάχνουμε τα κατηγορήματα που θέλουμε να αποθηκέυσουμε
#αρχικά σε μια λίστα με το όνομα literals
literals = []
movie_score = {}
for row in data.itertuples(index=True, name='Pandas'):
  
  #___simple attributes___:                                      # διαβάζουμε όλες τις στείλες του πίνακα εκτός αυτών οι οποίες δεν έχουν στα κελιά τους μία απλή τιμή π.χ. plot_keywords, και καλούμε για αυτά την clean_text

  movie_title = clean_text(getattr(row, 'movie_title'))
  budget = clean_text(str(getattr(row,'budget')))
  homepage = clean_text(getattr(row,'homepage'))
  id = clean_text(str(getattr(row,'id')))
  language = clean_text(getattr(row,'language'))
  original_title = clean_text(getattr(row,'original_title'))
  overview = clean_text(getattr(row,'overview'))
  popularity= clean_text(str(getattr(row,'popularity')))
  release_date= clean_text(getattr(row,'release_date'))
  gross= clean_text(str(getattr(row,'gross')))
  duration= clean_text(str(getattr(row,'duration')))
  status = clean_text(getattr(row,'status'))
  tagline = clean_text(getattr(row,'tagline'))
  movie_title = clean_text(getattr(row, 'movie_title'))
  vote_average= clean_text(str(getattr(row,'vote_average')))
  num_voted_users=clean_text(str(getattr(row,'num_voted_users')))
  title_year=clean_text(str(getattr(row,'title_year')))
  country = clean_text(getattr(row,'country'))   
  director= clean_text(getattr(row,'director_name'))
  actor_1_name = clean_text(getattr(row,'actor_1_name'))
  actor_2_name = clean_text(getattr(row,'actor_2_name'))
  actor_3_name = clean_text(getattr(row,'actor_3_name'))

  #insert into literals: distinction between strings and integers:                      # για την εισαγωγή των κανόνων στην prolog Θα πρέπει να υπάρχει διάκριση σε strings και integers καθώς στην πρώτη περίπτωση χρησιμοποιούμε τα σύμβολα: ' '

  #strings:

  if homepage !="UNK":
    literals.append("homepage('"+ movie_title +"','" + homepage + "')")
  if language !="UNK":
    literals.append("language('"+ movie_title +"','" + language + "')")
  if original_title !="UNK":
    literals.append("original_title('"+ movie_title +"','" + original_title + "')")
  if overview !="UNK":
    literals.append("overview('"+ movie_title +"','" + overview + "')")
  if release_date !="UNK":
    literals.append("release_date('"+ movie_title +"','" + release_date + "')")
  if status !="UNK":
    literals.append("status('"+ movie_title +"','" + status + "')")
  if tagline !="UNK":
    literals.append("tagline('"+ movie_title +"','" + tagline + "')")
  if movie_title !="UNK":
    literals.append("movie_title('"+ movie_title +"','" + movie_title + "')")
  if country !="UNK":
    literals.append("country('"+ movie_title +"','" + country + "')")
  if director !="UNK":
    literals.append("director_name('"+ movie_title +"','" + director + "')")
  if actor_1_name !="UNK":
    literals.append("actor_1_name('"+ movie_title +"','" + actor_1_name + "')")
  if actor_2_name !="UNK":
    literals.append("actor_2_name('"+ movie_title +"','" + actor_2_name + "')")
  if actor_3_name !="UNK":
    literals.append("actor_3_name('"+ movie_title +"','" + actor_3_name + "')")

  #integers:

  if budget !="UNK":
    literals.append("budget('"+ movie_title +"'," + budget + ")")
  if id !="UNK":
    literals.append("id('"+ movie_title +"'," + id + ")")
  if popularity !="UNK":
    literals.append("popularity('"+ movie_title +"'," + popularity + ")")
  if gross !="UNK":
    literals.append("gross('"+ movie_title +"'," + gross + ")")
  if duration !="UNK":
    literals.append("duration('"+ movie_title +"'," + duration + ")")
  if vote_average !="UNK":
    literals.append("vote_average('"+ movie_title +"'," + vote_average + ")")
  if num_voted_users !="UNK":
    literals.append("num_voted_users('"+ movie_title +"'," + num_voted_users + ")")
  if title_year !="UNK":
    literals.append("title_year('"+ movie_title +"'," + title_year + ")")

    
 #___list of attributes seperated by "|"___:                                                  # διαχειριζόμαστε τώρα τις στήλες με πολλαπλές τιμές

  for genre in getattr(row, 'genres').split("|"):
    literals.append("genre('"+ movie_title +"','"+ clean_text(genre) +"')")
  for keyword in getattr(row, 'plot_keywords').split("|"):
    literals.append("plot_keyword('"+ movie_title +"','"+ clean_text(keyword) +"')")

 #___list of attributes in dictionary form___:                                                # διαχειριζόμαστε τώρα τις στήλες με τιμές σε μορφή λεξικού

  country_dict=ast.literal_eval(getattr(row, 'production_countries'))   #convert string to dictionary list
  if (country_dict!=[]):
    for c in country_dict:
      literals.append("production_country('"+movie_title+"','"+clean_text(c['iso_3166_1'])+"','"+clean_text(c['name'])+"')")

  company_dict=ast.literal_eval(getattr(row, 'production_companies')) 
  if (company_dict!=[]):
    for c in company_dict:
      literals.append("production_company('"+movie_title+"','"+clean_text(c['name'])+"',"+clean_text(str(c['id']))+")")

  spoken_languages_dict=ast.literal_eval(getattr(row, 'spoken_languages')) 
  if (spoken_languages_dict!=[]):
    for s in spoken_languages_dict:
      literals.append("spoken_language('"+movie_title+"','"+clean_text(s['iso_639_1'])+"','"+clean_text(s['name'])+"')")

#Η Prolog θέλει τα κατηγορήματά της με την σειρά 
literals.sort()
for literal in literals:
  prolog.assertz(literal)               # εισάγουμε τα κατηγορήματα στην Prolog
````

Όσο ανεβαίνουμε επίπεδα ομοιότητας σε καθένα από αυτά τα if-else clauses προστίθενται διαφορετικοί κανόνες με αποτέλεσμα το σύνολο κανόνων του κάθε clause να γίνεται ολοένα και πιο αυστηρό. ##Δημιουργία simple recommender με βάση τα κατηγορήματα find_similar

Με χρήση των κατηγορημάτων find_similar που κατασκευάστηκαν παραπάνω, δημιουργείται ο simple recommender, που παίρνει παίρνει ως παράμετρο το όνομα της ταινίας με βάση της οποίας θα γίνουν οι προτάσεις, ως εξης:

Αρχικοποιείται η λίστα answers με τις ταινίες που θα προταθούν, καθώς και ένα dictionary, το movies_compatability, το οποιο αντιστοιχίζει την κάθε προτεινόμενη ταινία με το επίπεδο ομοιότητάς της (αν μια ταινία Χ έχει βρεθεί πρώτα από το find_similar_8 θα ισχύει ότι movies_compatibility[Χ] = 8).

Ξεκινώντας από το μεγαλύτερο επίπεδο ομοιότητας 10 και σταδιακά μειώνοντας κατά 1 φτάνουμε μέχρι το επίπεδο 1. Για καθένα από τα επίπεδα αυτα βρίσκουμε με ένα query τις παρόμοιες με την ταινία που δώθηκε ως παράμετρος ταινίες για το αντίστοιχο επίπεδο ομοιότητας, ελέγχουμε αν έχουμε ξαναβρει την ταινία αυτή είτε στο ίδιο ειτε σε κάποιο μεγαλύτερο επίπεδο ομοιότητας. Αν την έχουμε ξαναβρεί την αγνοούμε, μιας και έχει μπει ήδη πιο ψηλά στην λίστα προτάσεων, αν πό την άλλη η ταινία δεν έχει ξαναβρεθεί την προσθέτουμε σε ένα βοηθητικό συνολο που χρησιμοποιούμε μόνο στο τρέχν επίπεδο και αποθηκεύουμε στο movies_compatibility το επίπεδο ομοιότητας της για να χρησιμοποιηθεί παρακάτω. 

Μόλις τελειώσουν όλες οι ταινίες του επιπέδου οι προτάσεις για το επίπεδο αυτό βρίσκονται στο βοηθητικό σύνολο (s10, s9, ...). Εφόσον όλες οι ταινίες του συνόλου αυτού ανήκουν στο ίδιο επίπεδο ομοιότητας δεν μας πειράζει η σχετική τους σειρά, άρα τις ταξινομούμε αλφαβητικά για να μην εμφανίζονται τυχαία και τις τοποθετούμε στην ουρά της λίστας answers που θα επιστρέψουμε στο τέλος της συνάρτησης.

Εφόσον εκτελούμε τα επίπεδα ομοιότητας ένα ένα με ανάποδη σειρά και τοποθετούμε τις προτάσεις του κάθε επιπέδου στην ουρα της λίστα answers, όλες οι προτάσεις μεγαλύτερων επιπέδων θα προηγούνται γιατί θα έχουν ήδη τοποθετηθεί στο answers σε προηγούμενο χρόνο. Με τον τρόπο αυτό διασφαλίζεται ότι στις προτάσεις που γίνονται οι ταινίες εμγανίζονται με φθίνουσα σειρά ομοιότητας.


Σημειώνουμε ότι η συνάρτηση simple_recommender που ακολουθεί τροποποιήθηκε ώστε να επιστρέφει όχι μόνο την λίστα παρόμοιων ταινιών, αλλά και ένα λεξικό που περιλαμβάνει τον βαθμό ομοιότητας (1-10) για καθεμία από τις ταινίες της λίστας.

````python
def simple_recommender(movie):
  
    movies_compatibility = {}             # dictionary shows the similarity level of a recommended movie
    answers = []


    s10 = set()                           # helper distinct set where the similar movies of level 10 that we will recommended will be temporarily stored
    q = prolog.query("find_similar_10('" + movie +"',M)")     # find similar movies (of level 10) with the movie that was passed as an argument
    for soln in q:                        # for each similar movie that was found in the above query q
        m = soln['M']                   
        if m not in s10:                  # if found the movie has not been found before 
            s10.add(soln['M'])            # add the movie to the distinct helper set
            movies_compatibility[m]=10    # also add the movie and its corresponding similarity_level (here is 10) in the movies_compatibility dictionary
    q.close()                             # close the query to ensure there will be no nested query errors 
    answers10 = list(sorted(s10))         # sort the level-10 similar movies that remained in the helper set. We can do this because they have the same similarity level
    answers.extend(answers10)             # add the sorted set to the tail of the answers, which is the list of recommended movies we want to return in the end

    s9 = set()
    q = prolog.query("find_similar_9('" + movie +"',M)")
    for soln in q:
        m = soln['M'] 
        if m not in s9 and m not in s10 : # if the movie has not been found in the same or in a higher similarity level before
            s9.add(soln['M'])             # add it to the helper set s9 that will end up being the recommended movies with similarity of level 9
            movies_compatibility[m]=9
    q.close()
    answers9 = list(sorted(s9))
    answers.extend(answers9)              # add the sorted helper set s9 to the tail of the answers, after answers of level 10 

    s8 = set()
    q = prolog.query("find_similar_8('" + movie +"',M)")
    for soln in q:
        m = soln['M'] 
        if m not in s8 and m not in s9  and m not in s10:
            s8.add(soln['M'])
            movies_compatibility[m]=8
    q.close()
    answers8 = list(sorted(s8))
    answers.extend(answers8)

    s7 = set()
    q = prolog.query("find_similar_7('" + movie +"',M)")
    for soln in q:
        m = soln['M'] 
        if m not in s7 and m not in s8  and m not in s9 and m not in s8 and m not in s10:
            s7.add(soln['M'])
            movies_compatibility[m]=7
    q.close()
    answers7 = list(sorted(s7))
    answers.extend(answers7)

    s6 = set()
    q = prolog.query("find_similar_6('" + movie +"',M)")
    for soln in q:
        m = soln['M'] 
        if m not in s6 and m not in s7 and m not in s8 and m not in s9 and m not in s10:
            s6.add(soln['M'])
            movies_compatibility[m]=6
    q.close()
    answers6 = list(sorted(s6))
    answers.extend(answers6)

    s5 = set()
    q = prolog.query("find_similar_5('" + movie +"',M)")
    for soln in q:
        m = soln['M'] 
        if m not in s5 and m not in s6 and m not in s7 and m not in s8 and m not in s9 and m not in s10:
            s5.add(soln['M'])
            movies_compatibility[m]=5
    q.close()
    answers5 = list(sorted(s5))
    answers.extend(answers5)

    s4 = set()
    q = prolog.query("find_similar_4('" + movie +"',M)")
    for soln in q:
        m = soln['M'] 
        if m not in s4 and m not in s5 and m not in s6 and m not in s7 and m not in s8 and m not in s9 and m not in s10:
            s4.add(soln['M'])
            movies_compatibility[m]=4
    q.close()
    answers4 = list(sorted(s4))
    answers.extend(answers4)

    s3 = set()
    q = prolog.query("find_similar_3('" + movie +"',M)")
    for soln in q:
        m = soln['M'] 
        if m not in s3 and m not in s4  and m not in s5 and m not in s6 and m not in s7 and m not in s8 and m not in s9 and m not in s10:
            s3.add(soln['M'])
            movies_compatibility[m]=3
    q.close()
    answers3 = list(sorted(s3))
    answers.extend(answers3)

    s2 = set()
    q = prolog.query("find_similar_2('" + movie +"',M)")
    for soln in q:
        m = soln['M'] 
        if m not in s2 and m not in s3 and m not in s4  and m not in s5 and m not in s6 and m not in s7 and m not in s8 and m not in s9 and m not in s10:
            s2.add(soln['M'])
            movies_compatibility[m]=2
    q.close()
    answers2 = list(sorted(s2))
    answers.extend(answers2)

    s1 = set()
    q = prolog.query("find_similar_1('" + movie +"',M)")
    for soln in q:
        m = soln['M'] 
        if m not in s1 and m not in s2 and m not in s3 and m not in s4  and m not in s5 and m not in s6 and m not in s7 and m not in s8 and m not in s9 and m not in s10:
            s1.add(soln['M'])
            movies_compatibility[m]=1
    q.close()
    answers1 = list(sorted(s1))
    answers.extend(answers1)

    return answers,movies_compatibility     # return the recommended movies list, as well as the movies_compatibility dictionary, which will be used in part 3
````

*   Τροποποίηθηκαν επίσης τα βάρη score_weights. Σύμφωνα με τα επίπεδα ομοιότητας που έχουμε ορίσει θεωρούμε ότι κάθε επίπεδο έχει διαφορετικό βάρος και πως σε μεγαλύτερα επίπεδα έχει μεγαλύτερη αξία να περάσει μια ταινία σε μεγαλύτερο επίπεδο ομοιότητας. Δηλαδή θεωρήθηκε ότι αν μια ταινία περάσει από το επίπεδο 7 στο επίπεδο 8 είναι πολύ πιο όμοια με την ζητούμενη, άρα θα πρέπει να κερδίσει περισσότερο βάρος από ότι όταν θα περάσει από το επίπεδο 1 στο επίπεδο 2. Για αυτό τον λόγο τα βάρη δεν έχουν οριστεί με σταθερό βήμα. Συγκεκριμένα :
      *    Για επίπεδα ομοιότητας από 1 μέχρι 4 οι ταινίες που επιστρέφονται είναι σχετικά γενικές. Οπότε δίνουμε μια πολύ μικρή τιμή βάρους 0,5 για τις πολύ γενικές ταινίες επιπέδου 1 και όσο ανεβαίνουμε επίπεδο οι ταινίες γίνονται ελαφρώς πιο συγκεκριμένες, άρα βάζουμε ένα μικρό βήμα στην αύξηση των βαρών, αξίας 0,5 .
      *    Φτάνοντας στα επίπεδα 5 και 6 προστιθενται αρκετοί κανόνες, άρα οι ταινίες έχουν αρχίσει να γίνονται πιο συγκεκριμένες, για αυτό και το βήμα αύξησης βαρών τώρα γίνεται μεγαλύετερο και ίσο με 1.
      *    Τέλος για τα επίπεδα 7, 8 και 9 το σύνολο των κανόνων έχει γίνει αρκετά μεγάλο, οπότε οι ταινίες είναι πλέον πολύ όμοιες με την ζητούμενη και επιπλέον είναι πολύ πιο δύσκολο μια ταινία να φτάσει μέχρι τα επίπεδα αυτά και να περάσει από το ένα επίπεδο στο άλλο. Για αυτό και τα βάρη τους έχουν μεγαλύτερες τιμές, αλλά και η αξία του βήματος μεταξύ των επιπέδων αυτών έχει μεγαλύτερη αξία, ίση με 1,5 .
      *     Για το επίπεδο 10 όλα τα if-else clauses ενωποιούνται και πρέπει όλοι οι κανόνες να ικανοποιούνται ταυτόχρονα. Επομένως για τον λόγο αυτό θεωρούμε ότι το επίπεδο ομοιότητας αυτό έχει πολύ μεγαύτερο βάρος από όλα τα προηγούμενα του και για αυτό παιρνει την μεγαλύτερη τιμή με απόσταση από το προηγούμενο επίπεδο ομοιότητας 9 ίση με 2,5 .

*  Η τελευταία τροποποίηση που έγινε είναι πως για να βρούμε το επίπεδο ομοιότητας της κάθε ταινίας χρησιμοποιείται το movies_compatibility που ορίστηκε στην παραπάνω συναρτηση simple recommender, το οποίο εδώ έχει όνομα μεταβλητής similarity_dictionary. Επομένως για να βρεθεί το βάρος που αντιστοιχεί κάθε φορά στο συγκεκριμένο επίπεδο ομοιότητας εκτελείται το :  score_weights[ similarity_dict[similar_movie] ] .



````python
from tqdm.notebook import tqdm
from sklearn.metrics import precision_score, recall_score, f1_score
import numpy as np
import random

rating_weights = {0: -1, 1: -0.5, 2:0, 3:0.4, 4:0.5, 5:1}
score_weights = {1: 0.5, 2: 1, 3: 1.5, 4: 2, 5: 3, 6: 4, 7: 5.5, 8: 7, 9: 8.5, 10: 11} 

#score_weights = {i:i + 1 for i in range(10)} # ανάλογα με τα επίπεδα ομοιότητας που έχουν οριστεί στην simple_recommender
#rating_weights = {0: -1.5, 1: -1, 2:-0.5, 3:0.5, 4:1, 5:1.5}

def train_recommender(ratings, rating_weights, score_weights, number_of_movies = 10):
    """
    Στην συνάρτηση αυτή μπορούμε να ορίζουμε ποιο υποσύνολο των αξιολογήσεων θα χρησιμοποιήσουμε για το train μαζί με τα βάρη ομοιότητας και σκορ
    Σε συνδυασμό με τον αριθμό των ταινιών που θέλουμε να χρησιμοποιήσουμε σαν σύνολο δεδομένων π.χ. 10 από τις 100 ή 3 από τις 100 κ.ο.κ
    Αν θέλουμε να χρησιμοποιήσουμε όλες τις ταινίες σαν training set τότε ορίζουμε το number_of_movies = - 1
    """

    if number_of_movies > len(ratings):
        number_of_movies = len(ratings)


    if number_of_movies != -1:
        indexes = random.sample(range(len(ratings)), number_of_movies)
        ratings = ratings.iloc[indexes]

    movie_score = {}
    for row in tqdm(ratings.itertuples(index=True, name='Pandas')):
        movie = clean_text(getattr(row, 'movie_title'))
        rating = getattr(row, 'rating')

        similar_movies, similarity_dict = simple_recommender(movie)

        for similar_movie in similar_movies:
            if similar_movie not in movie_score:
                movie_score[similar_movie] = rating_weights[int(rating)] * score_weights[similarity_dict[similar_movie]]
            else:
                movie_score[similar_movie] += rating_weights[int(rating)] * score_weights[similarity_dict[similar_movie]] #το weight θα το ορίσετε ανα επίπεδο ομοιότητας οι πολύ όμοιες ταινίες θα έχουν μεγαλύτερο βάρος
    return movie_score


#αυτό είναι ένα παράδειγμα για το πως θα μπορούσε να υλοποιήθει η predict
#έχουμε ορίσει ότι μια ταινία θα έπρεπε να είναι προτεινόμενη αν είχε σκορ > 0

def predict_example(ratings, movie_score):
    real, pred = [], []
    for i, row in enumerate(ratings.itertuples(index=True, name='Pandas')):
        movie = clean_text(getattr(row, 'movie_title'))
        rating = getattr(row, 'rating')

        if movie in movie_score:                        #αν έχουμε σχηματίσει βαθμολογία για την ταινία αυτή
            pred.append(int(movie_score[movie] > 0))    #heuristic για το αν μια ταινία είναι προτεινόμενη
            real.append(int(rating > 3))                # έτσι ορίζουμε ότι μια ταινία θα έπρεπε να είναι προτεινόμενη
            #η συνθήκη αυτή δεν μπορεί να αλλάξει
        else:                                           #δεν μπορούμε να προτείνουμε κάτι για το όποιο δεν έχουμε σχηματίσει εικόνα
            pred.append(0)
            real.append(int(rating > 3))

    return real, pred


def get_metrics(real, pred):
    metrics = {}
    metrics["precision"] = precision_score(real, pred)
    metrics["recall"] = recall_score(real, pred)
    metrics["f1"] = f1_score(real, pred)
    return metrics
````

Η τιμές του F1 score, ως μία εξισορρόπηση των δύο παραπάνω μετρικών, βρίσκεται ανάμεσα στις τιμές αυτών και ξεκινά από 57% περίπου για μικρό σύνολο εκπαίδευσης, φτάνοντας για περισσότερες ταινίες το 67% περίπου. Πρόκειται για μία τιμή η οποία σίγουρα δεν είναι βέλτιστη, δείχνει όμως πως το σύστημα έχει αρκετά καλές επιδόσεις.

Αναλύοντας περαιτέρω τα αποτελέσματα των τριών μετρικών, συμπεραίνουμε ότι ο recommender μας είνα αρκετά "ανεκτικός" στο ποιες ταινίες θεωρεί όμοιες. Αυτό έχει ως αποτέλεσμα όλες οι όμοιες ταινίες να βρίσκονται σωστά (υψηλό recall), αλλά και περίπου άλλες τότσες ταινίες, οι οποίες ίσως δεν μοιάζουν τόσο, να θεωρούνται από τον recommender παρόμοιες (χαμηλότερο precision).

Θα μπορούσαμε ίσως να βελτιώσουμε το precision κάνοντας τους κανόνες find similar πιο αυστηρούς, ώστε να βρίσκουν λιγότερες και περισσότερο όμοιες ταινίες, ή θα μπορούσαμε να τροποποιήσουμε τα score_weights και rating_weights, ώστε οι ταινίες χαμηλότερου επιπέδου ομοιότητας να έχουν λιγότερο βάρος ή οι χαμηλές βαθμολογίες να μετράνε περισσότερο αρνητικά. Όλες οι επιλογές που αναφέρθηκαν θα είχαν ως αποτέλεσμα οι προτεινόμενες ταινίες να γίνουν λιγότερες. Κάτι τέτοιο θα μπορούσε να επιδράσει θετικά στην μετρική presicion, καθώς είναι πιθανό πολλές από τις ταινίες που θα απορρίπτονταν πλέον από το σύνολο να ήταν ταινίες που δεν έπρεπε εξ αρχής να προταθούν. Από την άλλη όμως, δεν είναι σίγουρο πως το αποτέλεσμα στο presicion θα ήταν θετικό: Αν ανάμεσα στις ταινίες που απορρίπτονταν βρίσκονταν ταινίες οι οποίοι θα έπρεπε να είχαν προταθεί, τότε η τιμή του presicion θα έπεφτε. Επιπλέον, στην περίπτωση αυτή πιθανόν να μειωνόταν η τιμή του recall αφού, όπως αναφέραμε, είναι πιθανό κάποιες από τις ταινίες που απορρίφθηκαν να έπρεπε εν τέλει να προταθούν.
Πράγματι, μετά από σχετικές δοκιμές, διαπιστώσαμε ότι εφαρμόζοντας τις παραπάνω ιδέες δεν υπήρξε σημαντική βελτίωση του precision, ενώ το recall μειώθηκε αισθητά. Αποφασίστηκε λοιπόν, εφόσον η τιμή του recall είναι εξαιρετικά υψηλή και η τιμή του precision είναι αποδεκτά καλή με την παρούσα υλοποίηση, να μην γίνει κάποια από τις προαναφερθείσες τροποποιήσεις. Αξίζει να σχολιάσουμε ότι ένας παράγοντας που μπορεί εμποδίζει την βελτίωση της συνολικής επίδοσης του συστήματός μας καθώς το κάνουμε πιο αυστηρό είναι η φύση αυτού: Πρόκειται για ένα σύστημα που θέτει ορισμένους κανόνες ομοιότητας οι οποίες είναι πολύ απόλυτοι και όμοιοι για όλους τους χρήστες. Όμως, καθώς κάθε χρήστης είναι διαφορετικός, μπορεί να είναι διαφορετικά και τα κριτήριά του: Ένας χρήστης θα μπορούσε να είναι fan ενός συγκεκριμένου ηθοποιού ή σκηνοθέτη και να αξιολογεί θετικά όλες τις ταινίες αυτού, ενώ ένας άλλος μπορεί να μην ενδιαφέρετια για κάτι τέτοιο, να έχει όμως μια σαφή προτίμηση σε ταινίες εποχής. Έτσι, οι προτάσεις που αναμένει να δει κάθε χρήστης διαφέρουν. Κάνοντας τον recommender πιο "ανεκτικό", μπορούμε να κάνουμε προτάσεις που ταιριάζουν σε πολλά διαφορετικά προφίλ χρηστών. Αυτό σημαίνει πως θα προτείνουμε πολλές ταινίες που θα αρέσουν στον κάθε χρήστη, αναπόφευκτα όμως θα προτείνουμε και άλλες που δεν του αρέσουν. Αντίθετα, κάνοντας το σύστημα πιο συγκεκριμένο, μπορεί για κάποιον χρήστη αυτό να είναι θετικό, και να περιορίσουμε τις προτάσεις που δεν του ταιριάζουν, είναι όμως πολύ εύκολο να παραβλέψουμε κάποια κριτήρια που θέτει ένας χρήστης για το ποια ταινία θα του αρέσει, και να προτείνουμε λιγότερες ή λάθος ταινίες σε αυτόν.

Παρατηρείται ότι η τιμή του recall μεγιστοποιείται στο 99% ενώ του precision στο 51% , ενώ η συνδυαστική τιμή f1 μεγιστοποιείται στην τιμή 67% . Παρατηρούμε επίσης ότι οι τιμές αυτές προσεγγίζονται πολύ γρήγορα, για μικρά μεγέθη υποσυνόλων εκπαίδευσης. Ήδη για 10 ταινίες οι τιμή του recall έχει φτάσει το 93% και του precision το 50% ενώ η συνδυαστική τους τιμή f1 ανέρχεται στο 65% , για μέγεθος υποσυνόλου εκπαίδευσης 20 ταινιών το recall έχει ανέβει στο 98% , το Precision είναι της τάξης του 50% και η συνδυαστική τους τιμή f1 έχει φτάσει ήδη στο 67% . Για υποσύνολα μεγέθους μεγαλύτερου των 20 ταινιών η τιμή των recall, precision και η συνδυαστική τιμή f1 φαίνεται να σταθεροποιούται. 

Το γεγονός ότι φτάνουμε σχεδόν στις μέγιστες τιμές recall, precision και f1 για τόσο μικρό μέγεθος υποσυνόλου επαίδευσης (20 ταινίες ή ακόμα και οι 10 ταινίες) μας οδηγεί στο συμπρέρασμα ότι, ο recommender που εν τέλει δημιουργήσαμε, είναι ιδιαίτερα κατάλληλος για νέους χρήστες. Οι νέοι χρήστες δεν έχουν βαθμολογήσει πολλές ταινίες, επομένως δεν έχουμε πολλά δεδομένα για αυτούς, ωστόσο παρατηρείται ότι ο recommender μας ακόμα και για πολύ μικρά training set, δηλαδή πολύ λίγα δεδομένα όσον αφορά τις προτιμήσεις του χρήστη, έχει καλή απόδοση και προτείνει σχεδόν όλες τις ταινίες που πρέπει. Προτείνει βέβαια και πολλές επιπλέον, κάτι τέτοιο θεωρούμε όμως αρκετά αποδεκτό για έναν νέο χρήση. Όσον αφορά παλαιότερους χρήστες, ο recommender εξακολουθεί να έχει αρκετά ικανονποιητική επίδοση (με F1 score 67% περίπου), δεν βελτιώνει όμως την ακρίβειά του σηματνικά σε σχέση με όταν ο χρήστης ήταν πιο νέος, γεγονός που σίγουρα δεν είναι βέλτιστο, καθώς ένας χρήστης μπορεί να είναι πιο δεκτικός σε λάθη του συστήματος στην αρχή, αναμένει όμως συνήθως να δει αισθητή βελτίωση στις προτάσεις που λαμβάνει, όσο μεγαλύτερη ανατροφοδότηση παρέχει στο σύστημα.
